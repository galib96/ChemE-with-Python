The vapor pressure of a pure liquid, written p∗, is a strong function of
temperature. To calculate the vapor pressure at a given temperature, T, it
is common to use Antoine’s equation:

<img src="Antoine Equation.JPG">

where A, B, and C are constants that can be looked up for different liquids.
Write a function that has A, B, C, and T (in ∘C) as inputs and returns the
vapor pressure.

Problem Source: Jeffrey J Heys, Chemical and Biomedical Engineering Calculations using Python, 2017 edition, Page 47

```python
def vapor_pressure_pure_liquid(T, A, B, C):
    """
    Return: Vapor pressure of a pure liquid at a given temperature. (mm Hg)
    T = Given temperature at which vapor pressure will be measured (in celsius)
    A, B, C = constants (varies for different liquid, need to look up the table)
    """
    # If we play with a equation a bit, we can see that vapor pressure P is
    # 10^(A-B/(T+C))
    # firstly, calculating the exponent
    exp = A-B/(T+C)

    # vapor pressure in mm Hg
    vapor_pressure = 10**(exp)
    return vapor_pressure
```

```python
# Test case:
# For Aniline, A = 7.32010, B = 1731.515 C = 206.049
liquid_name = "Aniline"
A = 7.32010
B = 1731.515
C = 206.049
# Calculating for multiple Temperatures
#temperature list
T_list = list(range(105,186,10))
# calculating vapor pressure for all the temperatures in the list
# we'll save the result in a list
vapor_pressure_dict = {}
for T in T_list:
    vp = round(vapor_pressure_pure_liquid(T,A,B,C),2)
    vapor_pressure_dict[T] = vp

import pandas as pd

vapor_pressure_df = pd.DataFrame(vapor_pressure_dict.items(), columns=['Temperature_C', 'Pressure_mmHg'])
vapor_pressure_df
```

```python
#visualizing the result from test case
import matplotlib.pyplot as plt

plt.scatter(vapor_pressure_df['Temperature_C'], vapor_pressure_df['Pressure_mmHg'])
plt.xlabel("Temperature (C)")
plt.ylabel("Vapor Pressure (mm Hg)")
plt.title("Vapor pressure of "+liquid_name+" with Temperature (Antoine equation)")

plt.show()
```
